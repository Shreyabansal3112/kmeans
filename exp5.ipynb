```
x=data.drop(['Unnamed: 5','variety'],axis=1)
y=data['variety']
print(x.shape)
print(y.shape)
```

```
tmp.median()
```

```
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.4,random_state=42)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```
km = KMeans(n_clusters=3,random_state=1)
km.fit(x_train)
```

```
km.labels_
```

```
km.cluster_centers_
```

```
x1 = data['sepal.length']
x2 = data['sepal.width']
x3 = data['petal.length']
x4 = data['petal.width']
y=km.labels_
colors = ['y','g','r']


markers = ['.', 'X','p','o']

for i , l in enumerate(km.labels_):
  plt.plot(x1[i],x2[i], color=colors[l],marker=markers[l])
```
